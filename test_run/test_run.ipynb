```python
# importing samples from legal bench and structuring the prompts
from datasets import load_dataset

dataset = load_dataset("nguha/legalbench", "privacy_policy_qa", split="test")

print(dataset[0])
```

```python
# structure prompts in json
import json

formatted_data = []

for i in range(25):
    sample = dataset[i]

    entry = {
        "input": f'Clause: {sample['text']}\nQuestion: {sample['question']}\nAnswer:',
        'target': sample['answer']
    }
    formatted_data.append(entry)

with open ("data/forget_set_prompts.json", "w") as final:
    json.dump(formatted_data, final, indent=4)
```
