# Test Run
### This is a small scale implementation of our project to ensure that it is actually feasible to complete within the deadline.
I will not be implementing ALKN in this test run; I will use a much simpler unlearning technique for this since the ALKN application is part of the major research in the project as a whole. I will load the LegalBench dataset and the comma_v0.1_1t model. Then, I will evaluate the performance of the chosen LegalBench tasks before and after unlearning. This should demonstrate how the rest of our project will go but on a smaller scale.
